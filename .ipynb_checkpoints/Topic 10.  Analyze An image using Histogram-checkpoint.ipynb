# Analyze An image using Histogram

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
blank_w = np.ones((500,500,3) , np.uint8 )*255 
blank_b = np.zeros((500,500,3) , np.uint8 )*255
```

```python
image = cv2.imread('eisa.jpeg')
# hist = cv2.calcHist([blank_w], [0] , None , [256], [0,256])
hist = cv2.calcHist([image], [0] , None , [256], [0,256])
plt.plot(hist)
plt.show()

# cv2.imshow("blank w" , blank_w)
cv2.imshow("blank_b" , image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the image in grayscale
image = cv2.imread('eisa.jpeg', cv2.IMREAD_GRAYSCALE)

# Calculate the histogram for grayscale
histogram = cv2.calcHist([image], [0], None, [256], [0, 256])

# Plot the histogram
plt.figure()
plt.title("Grayscale Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.plot(histogram)
plt.xlim([0, 256])
plt.show()
```

```python
# Load the image in color
image = cv2.imread('eisa.jpeg')

# Define color channels for plotting
colors = ('b', 'g', 'r')

# Plot histogram for each color channel
plt.figure()
plt.title("Color Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

for i, color in enumerate(colors):
    histogram = cv2.calcHist([image], [i], None, [256], [0, 256])
    plt.plot(histogram, color=color)
    plt.xlim([0, 256])

plt.show()
```

# histogram equlization 

```python
import cv2
import matplotlib.pyplot as plt

# Load the image in grayscale and color
image_gray = cv2.imread('eisa.jpeg', cv2.IMREAD_GRAYSCALE)
image_color = cv2.imread('eisa.jpeg')

# Apply histogram equalization on grayscale image
equalized_gray = cv2.equalizeHist(image_gray)

# For color image: Convert to YCrCb, equalize Y channel, and convert back to BGR
ycrcb = cv2.cvtColor(image_color, cv2.COLOR_BGR2YCrCb)
ycrcb[:, :, 0] = cv2.equalizeHist(ycrcb[:, :, 0])
equalized_color = cv2.cvtColor(ycrcb, cv2.COLOR_YCrCb2BGR)

# Display original and equalized images
plt.figure(figsize=(10, 5))

# Original grayscale image
plt.subplot(2, 2, 1)
plt.title("Original Grayscale")
plt.imshow(image_gray, cmap='gray')

# Equalized grayscale image
plt.subplot(2, 2, 2)
plt.title("Equalized Grayscale")
plt.imshow(equalized_gray, cmap='gray')

# Original color image
plt.subplot(2, 2, 3)
plt.title("Original Color")
plt.imshow(cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB))

# Equalized color image
plt.subplot(2, 2, 4)
plt.title("Equalized Color")
plt.imshow(cv2.cvtColor(equalized_color, cv2.COLOR_BGR2RGB))

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
      
img = cv2.imread('eisa.jpeg')
img = cv2.resize(img , (500,500))

gry = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)

hist = cv2.calcHist([img], [0],None , [255] , [0,255])
eq = cv2.equalizeHist(gry)


plt.plot(hist)
plt.show()

cv2.imshow("eq" , eq)
cv2.imshow("image" , img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Callback function to handle mouse events
def mouse_event(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        print(f"Left Button Clicked at ({x}, {y})")
        cv2.circle(img, (x, y), 5, (255, 0, 0), 4)  # Draw blue circle at click position
    elif event == cv2.EVENT_RBUTTONDOWN:
        print(f"Right Button Clicked at ({x}, {y})")
        cv2.circle(img, (x, y), 5, (0, 255, 0), 4)  # Draw green circle at click position
    elif event == cv2.EVENT_MBUTTONDOWN:
        print(f"Middle Button Clicked at ({x}, {y})")
        cv2.circle(img, (x, y), 5, (0, 0, 255), 4)  # Draw red circle at click position

# Create a blank image and window
img = np.zeros((512, 512, 3), np.uint8)
cv2.namedWindow("Image")

# Bind the mouse event to the window
cv2.setMouseCallback("Image", mouse_event)

while True:
    cv2.imshow("Image", img)
    if cv2.waitKey(1) & 0xFF == 27:  # Press 'Esc' to exit
        break

cv2.destroyAllWindows()
```
